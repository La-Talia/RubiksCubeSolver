```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("test/1.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.resize(img, (300, 300))
plt.imshow(img)
plt.show()
```

```python
segment = img[0:100, 0:100]
plt.imshow(segment)
plt.show()
```

```python
R=np.median(segment[:,:,0])
G=np.median(segment[:,:,1])
B=np.median(segment[:,:,2])

print(R,G,B)


```

```python
canvas = np.full((100, 100, 3), (R, G, B), dtype=int)
plt.imshow(canvas)
plt.show()
```

```python
C={'W':(255,255,255),
'Y':(255,255,0),
'O':(255,165,0),
'R':(255,0,0),
'G':(0,255,0),
'B':(0,0,255)}
```

```python
print(C['W'])
```

```python
D={}
for color in C.keys():
    distance=np.sqrt((C[color][0]-R)**2+(C[color][1]-G)**2+(C[color][2]-B)**2)
    D[color]=distance
for i in D:
    print(i, ':', D[i])
md=min(D, key=D.get)
print(md)
```
